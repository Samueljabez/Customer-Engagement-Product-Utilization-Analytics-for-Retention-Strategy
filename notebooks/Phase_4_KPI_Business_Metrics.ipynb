```python
import pandas as pd
import numpy as np
import json
import warnings
warnings.filterwarnings('ignore')

# Load engineered data
df = pd.read_csv('../data/data_with_engineered_features.csv')
print(f"Data loaded: {df.shape}")
```

# Phase 4: KPI Calculation and Business Metrics

## Business-Focused Analytics

This phase transforms analytical findings into actionable business metrics that executives and operations teams can use for strategic decision-making.

### KPI Categories
1. **Engagement Metrics** - Customer activity indicators
2. **Revenue Metrics** - Value and loyalty indicators
3. **Risk Metrics** - Churn prediction and at-risk segments
4. **Performance Metrics** - Channel and demographic performance
5. **Strategic Metrics** - Long-term health indicators

---

## Objective
Create executive dashboard KPIs that drive business decisions

## 1. Engagement Metrics

```python
print("=" * 80)
print("ENGAGEMENT KPIS")
print("=" * 80)

# 1.1 Active Member Ratio
active_ratio = df['IsActiveMember'].mean()
print(f"\n1.1 Active Member Ratio: {active_ratio:.2%}")
print(f"    Active Customers: {df['IsActiveMember'].sum():,}")
print(f"    Inactive Customers: {(df['IsActiveMember']==0).sum():,}")

# 1.2 Engagement Retention Ratio
active_retention = 1 - df[df['IsActiveMember']==1]['Exited'].mean()
inactive_retention = 1 - df[df['IsActiveMember']==0]['Exited'].mean()
engagement_retention_ratio = active_retention / inactive_retention
print(f"\n1.2 Engagement Retention Ratio: {engagement_retention_ratio:.2f}x")
print(f"    Active Retention: {active_retention:.2%}")
print(f"    Inactive Retention: {inactive_retention:.2%}")

# 1.3 Multi-Product Penetration
multi_product_pct = df['Multi_Product_Customer'].mean()
avg_products = df['NumOfProducts'].mean()
print(f"\n1.3 Multi-Product Penetration: {multi_product_pct:.2%}")
print(f"    Average Products per Customer: {avg_products:.2f}")
print(f"    Single Product Customers: {(df['NumOfProducts']==1).sum():,}")
print(f"    Multi-Product Customers: {(df['NumOfProducts']>=2).sum():,}")

# 1.4 Credit Card Adoption
credit_card_adoption = df['Has_Credit_Card'].mean()
print(f"\n1.4 Credit Card Adoption Rate: {credit_card_adoption:.2%}")
print(f"    Customers with Credit Cards: {df['Has_Credit_Card'].sum():,}")
```

## 2. Revenue & Value Metrics

```python
print("\n" + "=" * 80)
print("REVENUE & VALUE KPIS")
print("=" * 80)

# 2.1 Average Balance
avg_balance = df['Balance'].mean()
median_balance = df['Balance'].median()
total_balance = df['Balance'].sum()
print(f"\n2.1 Balance Metrics:")
print(f"    Total Customer Balances: ${total_balance:,.0f}")
print(f"    Average Balance: ${avg_balance:,.0f}")
print(f"    Median Balance: ${median_balance:,.0f}")

# 2.2 Premium Customer Segment
premium_customers = df['Premium_Customer'].sum()
premium_pct = df['Premium_Customer'].mean()
premium_avg_balance = df[df['Premium_Customer']==1]['Balance'].mean()
print(f"\n2.2 Premium Customer Metrics:")
print(f"    Premium Customers: {premium_customers:,} ({premium_pct:.2%})")
print(f"    Avg Premium Balance: ${premium_avg_balance:,.0f}")
print(f"    Premium Churn Rate: {df[df['Premium_Customer']==1]['Exited'].mean():.2%}")

# 2.3 High-Value Customer Retention
high_value_customers = df[df['Customer_Value_Score'] >= 60]
high_value_retention = 1 - high_value_customers['Exited'].mean()
print(f"\n2.3 High-Value Customer Retention:")
print(f"    High-Value Customers: {len(high_value_customers):,} ({len(high_value_customers)/len(df):.2%})")
print(f"    Retention Rate: {high_value_retention:.2%}")
print(f"    Value of Retained High-Value: ${high_value_customers[high_value_customers['Exited']==0]['Balance'].sum():,.0f}")
```

## 3. Churn & Risk Metrics

```python
print("\n" + "=" * 80)
print("CHURN & RISK KPIS")
print("=" * 80)

# 3.1 Overall Churn Rate
overall_churn = df['Exited'].mean()
total_churned = df['Exited'].sum()
total_retained = (df['Exited']==0).sum()
print(f"\n3.1 Overall Churn Metrics:")
print(f"    Total Churn Rate: {overall_churn:.2%}")
print(f"    Churned Customers: {total_churned:,}")
print(f"    Retained Customers: {total_retained:,}")

# 3.2 At-Risk Customers
at_risk = df[df['Churn_Risk_Score'] >= 60]
at_risk_churn = at_risk['Exited'].mean()
print(f"\n3.2 At-Risk Customer Segment:")
print(f"    Customers with High Risk Score: {len(at_risk):,} ({len(at_risk)/len(df):.2%})")
print(f"    Predicted Churn Rate: {at_risk_churn:.2%}")
print(f"    At-Risk Balance at Stake: ${at_risk['Balance'].sum():,.0f}")

# 3.3 High-Balance Inactive (Premium at Risk)
wealth_risk = df[(df['Balance'] > 100000) & (df['IsActiveMember'] == 0)]
wealth_risk_churn = wealth_risk['Exited'].mean()
print(f"\n3.3 Premium Inactive (Wealth Risk):")
print(f"    Customers: {len(wealth_risk):,}")
print(f"    Balance at Risk: ${wealth_risk['Balance'].sum():,.0f}")
print(f"    Churn Rate: {wealth_risk_churn:.2%}")
print(f"    Revenue Impact if Churned: ${wealth_risk[wealth_risk['Exited']==1]['Balance'].sum():,.0f}")

# 3.4 Segmental Churn Analysis
print(f"\n3.4 Churn by Customer Segment:")
for segment in df['Customer_Segment'].unique():
    seg_data = df[df['Customer_Segment'] == segment]
    seg_churn = seg_data['Exited'].mean()
    seg_count = len(seg_data)
    print(f"    {segment}: {seg_churn:.2%} ({seg_count:,} customers)")
```

## 4. Performance by Demographics

```python
print("\n" + "=" * 80)
print("DEMOGRAPHIC PERFORMANCE KPIS")
print("=" * 80)

# Geography Performance
print(f"\n4.1 Geographic Performance:")
for geo in df['Geography'].unique():
    geo_data = df[df['Geography'] == geo]
    print(f"\n    {geo}:")
    print(f"      Customers: {len(geo_data):,}")
    print(f"      Churn Rate: {geo_data['Exited'].mean():.2%}")
    print(f"      Avg Balance: ${geo_data['Balance'].mean():,.0f}")
    print(f"      Active Members: {geo_data['IsActiveMember'].mean():.2%}")

#  Gender Performance
print(f"\n4.2 Gender Performance:")
for gender in df['Gender'].unique():
    gender_data = df[df['Gender'] == gender]
    print(f"\n    {gender}:")
    print(f"      Customers: {len(gender_data):,}")
    print(f"      Churn Rate: {gender_data['Exited'].mean():.2%}")
    print(f"      Avg Age: {gender_data['Age'].mean():.1f}")
    print(f"      Active Members: {gender_data['IsActiveMember'].mean():.2%}")

# Age Groups
print(f"\n4.3 Age Group Performance:")
age_bins = [0, 30, 40, 50, 100]
age_labels = ['<30', '30-40', '40-50', '50+']
df['Age_Group'] = pd.cut(df['Age'], bins=age_bins, labels=age_labels, right=False)
for age_group in age_labels:
    age_data = df[df['Age_Group'] == age_group]
    if len(age_data) > 0:
        print(f"\n    {age_group}:")
        print(f"      Customers: {len(age_data):,}")
        print(f"      Churn Rate: {age_data['Exited'].mean():.2%}")
        print(f"      Active Members: {age_data['IsActiveMember'].mean():.2%}")
```

```python
# Create comprehensive KPI report
print("\n\n" + "=" * 80)
print("EXECUTIVE KPI DASHBOARD SUMMARY")
print("=" * 80)

kpi_report = {
    "OVERALL_HEALTH": {
        "Total_Customers": int(len(df)),
        "Churned_Customers": int(df['Exited'].sum()),
        "Overall_Churn_Rate": f"{overall_churn:.2%}",
        "Overall_Retention_Rate": f"{1-overall_churn:.2%}"
    },
    "ENGAGEMENT": {
        "Active_Member_Ratio": f"{active_ratio:.2%}",
        "Engagement_Retention_Ratio": f"{engagement_retention_ratio:.2f}x",
        "Multi_Product_Penetration": f"{multi_product_pct:.2%}",
        "Average_Products": f"{avg_products:.2f}",
        "Credit_Card_Adoption": f"{credit_card_adoption:.2%}"
    },
    "FINANCIAL": {
        "Total_Customer_Balance": f"${total_balance:,.0f}",
        "Average_Balance": f"${avg_balance:,.0f}",
        "Premium_Customers": f"{premium_customers:,}",
        "Premium_Customer_Pct": f"{premium_pct:.2%}",
        "High_Value_Customers": len(high_value_customers),
        "High_Value_Retention": f"{high_value_retention:.2%}"
    },
    "RISK": {
        "At_Risk_Customers": len(at_risk),
        "At_Risk_Pct": f"{len(at_risk)/len(df):.2%}",
        "Premium_Inactive": len(wealth_risk),
        "Premium_Inactive_Balance": f"${wealth_risk['Balance'].sum():,.0f}",
        "Premium_Inactive_Churn": f"{wealth_risk_churn:.2%}"
    }
}

# Save KPI report
json_path = '../data/kpi_report.json'
with open(json_path, 'w') as f:
    json.dump(kpi_report, f, indent=2)

print("\n\nKPI Report JSON:")
print(json.dumps(kpi_report, indent=2))
print(f"\n✓ KPI report saved to: {json_path}")
```
